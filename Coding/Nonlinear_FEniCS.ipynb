## Euler's Elastica Equation

\begin{align}
\frac{d^{2}\theta}{ds^2} + \lambda^2 sin(\theta) = 0,\\
\theta(0) = \theta(1) = 0
\end{align}

```python
from ngsolve import *
from ngsolve.meshes import Make1DMesh
from ngsolve.webgui import Draw
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
mesh = Make1DMesh(32, mapping=None, periodic=False)
lambd = Parameter(1.0)
```

```python
fes = H1(mesh, order=1, dirichlet="left|right")
```

```python
u = fes.TrialFunction()
v = fes.TestFunction()
```

```python
deflection = []
```

```python
for lambd in list(np.linspace(3*pi-0.01,4*pi,200)):
    a = BilinearForm(fes)
    a += (grad(u)*grad(v) - lambd**2*sin(u)*v)*dx

    gfsol = GridFunction(fes)
    gfsol.Set(-sin(x)) # initial guess
    solvers.Newton(a,gfsol,inverse="umfpack")
    index = int((len(gfsol.vec)+1)/2)
    deflection.append(gfsol.vec[index])
```

```python
Draw(gfsol,mesh, deformation=CF( (0,gfsol,0) ))
```

```python
x=np.linspace(3*pi-0.01,4*pi,200)
plt.plot(x,deflection,color='black',linestyle="-")
# plt.xlim([0, 4*pi])
plt.xlabel('$\lambda$')
plt.ylabel('Deflection at mid point')
```

```python
from ngsolve import *
from ngsolve.meshes import Make1DMesh
from ngsolve.webgui import Draw
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
mesh = Make1DMesh(500, mapping=None, periodic=False)
```

```python
myfunc = sin(pi*x)
```

```python
fes = H1(mesh, order=1, dirichlet="left|right")
u = GridFunction(fes)
u.Set(myfunc)
```

```python
x=np.linspace(0,1,100)
plt.plot(x,u(mesh(x)),color='black',label='q = -0.4',linestyle="-")
plt.ylim([0, 1.1])
```
