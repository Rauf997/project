```python
import matplotlib.pyplot as plt
```

```python
x = [ 150, 200, 250, 300,350,400,450,500,550,600,650 ]
y_1 = [ 42.125,23.704,15.173,10.538,7.742,5.928,4.684,3.794,3.135,2.635,2.245 ]
y_2 = [ 116.232,65.436,41.901,29.107,21.398,16.378,12.942,10.484,8.665,7.2818,6.205]
plt.plot(x, y_1, color="red", marker="o",  linestyle="-", label ='1st mode')
plt.plot(x, y_2, color="black", marker="o",  linestyle="-",label ='2nd mode')
plt.legend()
plt.xlabel("Radius[$\mu$ m]")
plt.ylabel("Critical Stress[MPa]")
```
